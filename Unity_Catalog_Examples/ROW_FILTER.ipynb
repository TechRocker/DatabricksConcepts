```python
%sql
CREATE OR REPLACE FUNCTION main.default.sales_country_filter(country STRING)
RETURNS BOOLEAN
RETURN IF(
    is_account_group_member(country),  -- user must belong to account group with same name as country
    TRUE,
    FALSE
);
```

```python
%sql
ALTER TABLE main.default.sales_summary
SET ROW FILTER main.default.sales_country_filter ON (COUNTRY);
```
