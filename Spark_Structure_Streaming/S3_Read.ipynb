```python
df = (
    spark.readStream.format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option(
        "cloudFiles.schemaLocation",
        "s3://dbxtrainingbucket/circuitbox/landing/operational_data/_schema/addresses",
    )
    .option("cloudFiles.inferColumnTypes", "true")
    # .option(
    #     "checkpointLocation",
    #     "s3://dbxtrainingbucket/circuitbox/landing/operational_data/_checkpoint/addresses/",
    # )
    .load("s3://dbxtrainingbucket/circuitbox/landing/operational_data/addresses")
)
# df.display()
```

```python
(
    df.writeStream.format("delta")
    .option(
        "checkpointLocation",
        "s3://dbxtrainingbucket/circuitbox/landing/operational_data/_checkpoint/addresses/",
    )
    # .trigger(processingTime="30 seconds")
    .trigger(once=True)
    .start(
        "s3://dbxtrainingbucket/circuitbox/deltalakehouse/operational_data/addresses"
    )
)
```
