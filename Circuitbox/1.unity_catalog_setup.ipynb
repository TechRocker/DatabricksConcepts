```python
%sql
-- Create catalog with managed location
CREATE CATALOG circuitbox
MANAGED LOCATION 's3://dbxtrainingbucket/circuitbox/catalog/'
COMMENT 'Managed catalog for Unity Catalog.'
```

```python
%sql
SHOW CATALOGS
```

```python
%sql
-- Set catalog context
USE CATALOG circuitbox;

-- Create schema with managed location
CREATE SCHEMA IF NOT EXISTS circuitbox.landing
MANAGED LOCATION 's3://dbxtrainingbucket/circuitbox/landing'
COMMENT 'Managed schema for Unity Catalog.';

-- Create schema with managed location
CREATE SCHEMA IF NOT EXISTS circuitbox.lakehouse
MANAGED LOCATION 's3://dbxtrainingbucket/circuitbox/lakehouse'
COMMENT 'Managed schema for Unity Catalog.'
```

```python
%sql
SHOW SCHEMAS
```

```python
%sql
-- Set catalog and schema context
USE CATALOG circuitbox;
USE SCHEMA landing;

-- Create volume
CREATE EXTERNAL VOLUME IF NOT EXISTS circuitbox.landing.operational_data
LOCATION 's3://dbxtrainingbucket/circuitbox/landing/operational_data'
COMMENT 'Volume for Unity Catalog.'
```
